# Cleaning the dataset

```python
dataset = data.drop(['Segment','Country','State','Region'], axis = 1)
```

```python
dataset.shape 
```

```python
dataset.head()
```

```python
#corelation matrix
corelation=dataset.corr()
print(corelation)
```

# Data Visualisation

```python
#heatmap of corelation matrix
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
sns.boxplot(x=dataset['Sales'])
```

```python
sns.boxplot(x=dataset['Discount'])
```

```python
sns.boxplot(x=dataset['Profit']) 
```

```python
sns.displot(dataset, x='Profit', kind='kde') #to know the shape of the distribution
```

```python
sns.displot(dataset,x='Quantity',hue='Sub-Category',kind='kde',fill='tree') 
#distribution for quantity of each item in sub-category
```

```python
sns.displot(dataset,x='Sales',hue='Sub-Category',kind='kde',fill='tree') 
#distribution for sales of each item in sub-category
```

```python
sns.displot(dataset,x='Discount',hue='Sub-Category',kind='kde',fill='tree')
#distribution for discount of each item in sub-category
```

```python
sns.displot(dataset,x='Profit',hue='Category',kind='kde',fill='tree') 
#distribution for profit of each item in category
```

```python
sns.relplot(x='Sales', y='Profit', hue='Sub-Category', data=dataset)
```
